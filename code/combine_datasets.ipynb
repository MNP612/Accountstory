* check for DMs in nDM dataset

```python
doubles = []
for (name, company) in zip(data_nDM['full_name'], data_nDM['job_company_name']):
    if name in list(data_DM['full_name']) and company in list(data_DM['job_company_name']):
        doubles.append(name)
        print(name, company)
print(len(doubles))
```
